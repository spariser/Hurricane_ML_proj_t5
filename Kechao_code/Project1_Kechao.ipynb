# Load SST Data (1980-2024)

```python
sst = xr.open_dataset("SST.nc")
sst
```

```python
# Parse storm time data (convert to np.datetime64)
storm_times = np.array([
    [np.datetime64(t.decode("utf-8")) if isinstance(t, bytes) else np.datetime64("NaT") for t in row]
    for row in storms["iso_time"].values
], dtype="datetime64[ns]")

# Parse SST time data (already in np.datetime64 format)
sst_times = sst["valid_time"].values  

# Extract latitude and longitude of storm tracks
storm_lats = storms["lat"].values
storm_lons = storms["lon"].values

# Create an empty SST variable array (pre-filled with NaN)
sst_values = np.full(storm_times.shape, np.nan)  

# Iterate through the storms dataset and match the closest SST data
for i in range(storms.dims["storm"]):  # Loop through each storm
    for j in range(storms.dims["date_time"]):  # Loop through each track point
        time = storm_times[i, j]
        lat = storm_lats[i, j]
        lon = storm_lons[i, j]

        if np.isnat(time) or np.isnan(lat) or np.isnan(lon):
            continue  # Skip invalid data points

        # Find the closest SST time index
        time_idx = np.argmin(np.abs(sst_times - time))
        nearest_time = sst_times[time_idx]

        # Retrieve SST value using the nearest-neighbor method
        sst_point = sst.sel(
            valid_time=nearest_time
        ).sel(
            latitude=lat, longitude=lon, method="nearest"
        )

        # Ensure the result is a numeric value
        sst_values[i, j] = float(sst_point.sst.values.item())  # `.item()` extracts scalar value

# Add the matched SST data to the storms dataset
storms["sst"] = (("storm", "date_time"), sst_values)
```

# Moments Calculation

```python
def map_background(ax, label=False, extent=[-100, 0, 0, 60]):
  # A helpder function for creating the map background.
  # INPUT:
  # "extent": corresponds to the location information of the showed map.
  # "label": boolean

  # OUTPUT:
  # Matplotlib AXES object
    
    ax.set_extent(extent, crs=ccrs.PlateCarree())  
    ax.coastlines()  
    ax.gridlines(draw_labels=label)  

    land = cfeature.NaturalEarthFeature('physical', 'land', '10m', 
                                        edgecolor='face', facecolor=cfeature.COLORS['land'])
    ocean = cfeature.NaturalEarthFeature('physical', 'ocean', '10m', 
                                         edgecolor='face', facecolor=cfeature.COLORS['water'])
    ax.add_feature(land, zorder=0)
    ax.add_feature(ocean, zorder=0)
    
    return ax

def get_lon_lat(storm):
  lon_lst = storm.lon.values
  lat_lst = storm.lat.values
  return lon_lst[~np.isnan(lon_lst)], lat_lst[~np.isnan(lat_lst)]

def plot_one_track(ax, storm, color='blue', all_track=True, marker_color='black'):
  # INPUT:
  # ax: Matplotlib axes object
  # storm: a Xarray DATASET object (this can be adjusted if desired)
  # all_track: plot the whole track or just the origin

  # OUTPUT:
  # None

  lon_lst, lat_lst = get_lon_lat(storm)
  year = storm.season.values
  if all_track:
    ax.plot(lon_lst, lat_lst, '-o', color=color, linewidth=1, markersize=1) # marker='.'
    # ax.plot(lon_lst[-1], lat_lst[-1], color=marker_color, marker='x', markersize=5)
  ax.plot(lon_lst[0], lat_lst[0], color=marker_color, marker='*', markersize=5)
  # ax.text(lon_lst[0], lat_lst[0]-2, str(storm.name.values)[2:-1], horizontalalignment='center')
    
def get_moments(storm):
  # A function to calculate the track moments given a storm
  # OUTPUT:
  # X-centroid, Y-centroid, X_var, Y_var, XY_var

  # Note that:
  # In this case, no weights are set. In other words, all weights are 1.
  # A weight variable would need to be added in order to explore other weights

  lon_lst, lat_lst = get_lon_lat(storm)
  # If the track only has one point, there is no point in calculating the moments
  if len(lon_lst)<= 1: return None
      
  # M1 (first moment = mean). 
  # No weights applied
  lon_weighted, lat_weighted = np.mean(lon_lst), np.mean(lat_lst)
    
  # M2 (second moment = variance of lat and of lon / covariance of lat to lon
  # No weights applied
  cv = np.ma.cov([lon_lst, lat_lst])
    
  return [lon_weighted, lat_weighted, cv[0, 0], cv[1, 1], cv[0, 1]]
```

```python
# It takes a moment to calculate the moments.
moment_lst = [get_moments(storms.sel(storm=i)) for i in range(storms.dims['storm'])
              if get_moments(storms.sel(storm=i))]
print(np.shape(moment_lst))
print(f"There are {storms.dims['storm']-np.shape(moment_lst)[0]} tracks that have length less than 1.")
```

# K-means Clustering

```python
sum_of_squares = []
K = range(1, 15)
for k in K:
    km = k_means(moment_lst, n_clusters=k)
    sum_of_squares.append(km[2])
plt.plot(K, sum_of_squares, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Inertia Values against the Number of Cluster k')
```

```python
storms = storms.assign_coords(cluster=("storm", km[1]))
unique_clusters = np.unique(storms["cluster"].values)

fig, axes = plt.subplots(2, 3, figsize=(15, 10), subplot_kw={'projection': ccrs.PlateCarree()})
axes = axes.flatten()

for i, cluster_id in enumerate(unique_clusters):
    ax = axes[i]
    ax = map_background(ax, extent=[-100, 0, 0, 60])  # 传入 ax
    cluster_storms = storms.where(storms.cluster == cluster_id, drop=True)
    
    for ind in range(cluster_storms.dims["storm"]):
        storm = cluster_storms.sel(storm=ind)
        plot_one_track(ax, storm, color="orange")

    ax.set_title(f"Cluster {cluster_id}")

plt.tight_layout()
plt.show()
```

# Maximum Wind Speed

```python
# Create a list to store max wind speed data for each cluster
cluster_wind_data = []
unique_clusters = np.unique(storms['cluster'].values)

for cluster in unique_clusters:
    cluster_data = storms.where(storms['cluster'] == cluster, drop=True)
    
    # Process each storm's max wind speed
    cluster_wind_list = []
    for storm_wind in cluster_data['usa_wind']:
        # Remove NaNs and find the max wind speed for this storm
        valid_wind = storm_wind[~np.isnan(storm_wind)]
        if len(valid_wind) > 0:
            cluster_wind_list.append(valid_wind.max().values)
    
    # Add to data only if the cluster_wind_list is not empty
    if cluster_wind_list:
        cluster_wind_data.append(cluster_wind_list)

# Make boxplots
plt.figure(figsize=(10, 6))
plt.boxplot(cluster_wind_data, labels=[f'Cluster {c}' for c in unique_clusters if len(cluster_wind_data[unique_clusters.tolist().index(c)]) > 0])
plt.title('Maximum Wind Speed Distribution by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Maximum Wind Speed (knots)')
plt.show()
```

# Correlation

```python
# Store the maximum SST and maximum wind speed for each cluster
cluster_sst_max = []
cluster_max_winds = []
unique_clusters = np.unique(storms['cluster'].values)

for cluster in unique_clusters:
    cluster_data = storms.where(storms['cluster'] == cluster, drop=True)
    
    max_wind_list = []
    max_sst_list = []
    
    for storm_sst, storm_wind in zip(cluster_data['sst'], cluster_data['usa_wind']):
        # Remove NaNs and find the maximum wind speed for each storm
        valid_wind = storm_wind[~np.isnan(storm_wind)]
        if len(valid_wind) > 0:
            max_wind_list.append(valid_wind.max().values)

        # Remove NaNs and find the maximum SST for each storm
        valid_sst = storm_sst[~np.isnan(storm_sst)]
        if len(valid_sst) > 0:
            max_sst_list.append(valid_sst.max().values)
    
    # Add data only if both SST and wind speed lists are not empty
    if max_wind_list and max_sst_list:
        cluster_sst_max.append(max_sst_list)
        cluster_max_winds.append(max_wind_list)

# Compute correlation and plot
plt.figure(figsize=(15, 8))
correlations = []

for i, cluster in enumerate(unique_clusters):
    if i >= len(cluster_sst_max) or i >= len(cluster_max_winds):
        continue  # Prevent index out of range

    sst_vals = np.array(cluster_sst_max[i])
    wind_vals = np.array(cluster_max_winds[i])

    if len(sst_vals) == len(wind_vals):  # Ensure matching data lengths
        corr, p_value = pearsonr(sst_vals, wind_vals)
        correlations.append((corr, p_value))

        # Scatter plot
        plt.subplot(2, int(np.ceil(len(unique_clusters) / 2)), i + 1)
        sns.regplot(x=sst_vals, y=wind_vals, scatter_kws={'alpha': 0.6}, line_kws={"color": "red"})
        
        significance = " (p < 0.05)" if p_value < 0.05 else " (n.s.)"
        plt.title(f'Cluster {cluster} (r={corr:.2f}{significance})')
        plt.xlabel('Maximum SST')
        plt.ylabel('Max Wind Speed')

plt.tight_layout()
plt.show()
```

```python
# Extract max SST and max wind speed for each storm
max_sst = []
max_wind = []

for storm_sst, storm_wind in zip(storms['sst'], storms['usa_wind']):
    # Remove NaNs and find max SST
    valid_sst = storm_sst[~np.isnan(storm_sst)]
    if len(valid_sst) > 0:
        max_sst.append(valid_sst.max().values)
    
    # Remove NaNs and find max wind speed
    valid_wind = storm_wind[~np.isnan(storm_wind)]
    if len(valid_wind) > 0:
        max_wind.append(valid_wind.max().values)

# Convert lists to arrays
max_sst = np.array(max_sst)
max_wind = np.array(max_wind)

# Ensure data lengths match
if len(max_sst) == len(max_wind):
    corr, p_value = pearsonr(max_sst, max_wind)

    # Scatter plot with regression line
    plt.figure(figsize=(8, 6))
    sns.regplot(x=max_sst, y=max_wind, scatter_kws={'alpha': 0.6}, line_kws={"color": "red"})
    plt.title(f'Max SST vs. Max Wind Speed (r={corr:.2f}, p={p_value:.4f})')
    plt.xlabel('Maximum SST')
    plt.ylabel('Maximum Wind Speed')
    plt.show()

    # Print correlation significance
    print(f'Pearson Correlation: r = {corr:.2f}, p = {p_value:.4f}')
else:
    print("Mismatch in data length, check preprocessing steps.")
```
