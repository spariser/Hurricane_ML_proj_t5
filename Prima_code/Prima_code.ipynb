```python
ws = xr.open_dataset("data/WS.nc")
ws
```

```python
# Parse storm time data (convert to np.datetime64)
storm_times = np.array([
    [np.datetime64(t.decode("utf-8")) if isinstance(t, bytes) else np.datetime64("NaT") for t in row]
    for row in storms["iso_time"].values
], dtype="datetime64[ns]")

# Parse WS time data (already in np.datetime64 format)
ws_times = ws["time"].values  

# Extract latitude and longitude of storm tracks
storm_lats = storms["lat"].values
storm_lons = storms["lon"].values

# Create an empty WS variable array (pre-filled with NaN)
ws_values = np.full(storm_times.shape, np.nan)  

# Iterate through the storms dataset and match the closest WS data
for i in range(storms.dims["storm"]):  # Loop through each storm
    for j in range(storms.dims["date_time"]):  # Loop through each track point
        time = storm_times[i, j]
        lat = storm_lats[i, j]
        lon = storm_lons[i, j]

        if np.isnat(time) or np.isnan(lat) or np.isnan(lon):
            continue  # Skip invalid data points

        # Find the closest WS time index
        time_idx = np.argmin(np.abs(ws_times - time))
        nearest_time = ws_times[time_idx]

        # Retrieve WS value using the nearest-neighbor method
        ws_point = ws.sel(
            time=nearest_time
        ).sel(
            lat=lat, lon=lon, method="nearest"
        )

        # Ensure the result is a numeric value; adjust the variable name if needed.
        ws_values[i, j] = float(ws_point.wind_shear_200_1000.values.item())  # assuming the variable is named "ws"

# Add the matched WS data to the storms dataset
storms["ws"] = (("storm", "date_time"), ws_values)
```

```python
def map_background(ax, label=False, extent=[-100, 0, 0, 60]):
  # A helpder function for creating the map background.
  # INPUT:
  # "extent": corresponds to the location information of the showed map.
  # "label": boolean

  # OUTPUT:
  # Matplotlib AXES object
    
    ax.set_extent(extent, crs=ccrs.PlateCarree())  
    ax.coastlines()  
    ax.gridlines(draw_labels=label)  

    land = cfeature.NaturalEarthFeature('physical', 'land', '10m', 
                                        edgecolor='face', facecolor=cfeature.COLORS['land'])
    ocean = cfeature.NaturalEarthFeature('physical', 'ocean', '10m', 
                                         edgecolor='face', facecolor=cfeature.COLORS['water'])
    ax.add_feature(land, zorder=0)
    ax.add_feature(ocean, zorder=0)
    
    return ax

def get_lon_lat(storm):
  lon_lst = storm.lon.values
  lat_lst = storm.lat.values
  return lon_lst[~np.isnan(lon_lst)], lat_lst[~np.isnan(lat_lst)]

def plot_one_track(ax, storm, color='blue', all_track=True, marker_color='black'):
  # INPUT:
  # ax: Matplotlib axes object
  # storm: a Xarray DATASET object (this can be adjusted if desired)
  # all_track: plot the whole track or just the origin

  # OUTPUT:
  # None

  lon_lst, lat_lst = get_lon_lat(storm)
  year = storm.season.values
  if all_track:
    ax.plot(lon_lst, lat_lst, '-o', color=color, linewidth=1, markersize=1) # marker='.'
    # ax.plot(lon_lst[-1], lat_lst[-1], color=marker_color, marker='x', markersize=5)
  ax.plot(lon_lst[0], lat_lst[0], color=marker_color, marker='*', markersize=5)
  # ax.text(lon_lst[0], lat_lst[0]-2, str(storm.name.values)[2:-1], horizontalalignment='center')
    
def get_moments(storm):
  # A function to calculate the track moments given a storm
  # OUTPUT:
  # X-centroid, Y-centroid, X_var, Y_var, XY_var

  # Note that:
  # In this case, no weights are set. In other words, all weights are 1.
  # A weight variable would need to be added in order to explore other weights

  lon_lst, lat_lst = get_lon_lat(storm)
  # If the track only has one point, there is no point in calculating the moments
  if len(lon_lst)<= 1: return None
      
  # M1 (first moment = mean). 
  # No weights applied
  lon_weighted, lat_weighted = np.mean(lon_lst), np.mean(lat_lst)
    
  # M2 (second moment = variance of lat and of lon / covariance of lat to lon
  # No weights applied
  cv = np.ma.cov([lon_lst, lat_lst])
    
  return [lon_weighted, lat_weighted, cv[0, 0], cv[1, 1], cv[0, 1]]
```

```python
# It takes a moment to calculate the moments.
moment_lst = [get_moments(storms.sel(storm=i)) for i in range(storms.dims['storm'])
              if get_moments(storms.sel(storm=i))]
print(np.shape(moment_lst))
print(f"There are {storms.dims['storm']-np.shape(moment_lst)[0]} tracks that have length less than 1.")
```

```python
sum_of_squares = []
K = range(1, 15)
for k in K:
    km = k_means(moment_lst, n_clusters=k)
    sum_of_squares.append(km[2])
plt.plot(K, sum_of_squares, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Inertia Values against the Number of Cluster k')
```

```python
km = k_means(moment_lst, n_clusters=4)
# print(km[0])

colors = ['black', 'red', 'blue', 'yellow', 'green', 'magenta']
labels = km[1]
fig, ax = plt.subplots(figsize=(15, 7), subplot_kw={"projection": ccrs.PlateCarree()})
ax = map_background(ax, label=True, extent=[-100, 0, 0, 60]) 

for k in range(len(moment_lst)):
    ax.plot(moment_lst[k][0], moment_lst[k][1], c=colors[labels[k]], marker='*')

plt.title('K-means clustering result, 6 clusters')
```

```python
storms = storms.assign_coords(cluster=("storm", km[1]))
unique_clusters = np.unique(storms["cluster"].values)

fig, axes = plt.subplots(2, 3, figsize=(15, 10), subplot_kw={'projection': ccrs.PlateCarree()})
axes = axes.flatten()

for i, cluster_id in enumerate(unique_clusters):
    ax = axes[i]
    ax = map_background(ax, extent=[-100, 0, 0, 60])  # 传入 ax
    cluster_storms = storms.where(storms.cluster == cluster_id, drop=True)
    
    for ind in range(cluster_storms.dims["storm"]):
        storm = cluster_storms.sel(storm=ind)
        plot_one_track(ax, storm, color="orange")

    ax.set_title(f"Cluster {cluster_id}")

plt.tight_layout()
plt.show()
```

```python
# Assign season as a coordinate and count the number of occurrences per season and cluster
storms = storms.assign_coords(season=storms["season"])
cluster_counts = storms["numobs"].groupby(["season", "cluster"]).count()
cluster_counts_df = cluster_counts.to_dataframe(name="count").reset_index()

# Get unique clusters
unique_clusters = sorted(cluster_counts_df['cluster'].unique())  # Ensure consistent order

# Create subplots with 2 rows and 3 columns
fig, axes = plt.subplots(2, 3, figsize=(15, 8), sharex=True, sharey=True)  # Shared axes for consistency
axes = axes.flatten()  # Flatten the 2D axes array for easy iteration

# Plot each cluster separately in a subplot
for i, cluster_id in enumerate(unique_clusters):
    cluster_data = cluster_counts_df[cluster_counts_df['cluster'] == cluster_id]

    axes[i].plot(cluster_data['season'], cluster_data['count'], marker='o', linestyle='-', label=f'Cluster {cluster_id}')
    axes[i].set_title(f'Cluster {cluster_id}')
    axes[i].legend()
    axes[i].grid(True)

# Adjust layout
for i in range(len(unique_clusters), len(axes)):  # Hide extra subplots if any
    fig.delaxes(axes[i])

fig.supxlabel('Season')
fig.supylabel('Count')
fig.suptitle('Number of Occurrences of Each Cluster Over Time', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Check if each storm experienced landfall (i.e., has at least one time step where landfall == 0)
storm_landfall_flag = (storms['landfall'] == 0).any(dim='date_time')

# Count the number of storms that experienced landfall for each cluster
cluster_landfall_counts = storm_landfall_flag.groupby(storms['cluster']).sum()

# Count the total number of storms in each cluster
cluster_total_counts = storms["numobs"].groupby('cluster').count()

# Calculate the landfall ratio for each cluster
cluster_landfall_ratio = (cluster_landfall_counts / cluster_total_counts).fillna(0)

# Convert to a DataFrame for better readability
landfall_stats_df = pd.DataFrame({
    'Landfall Count': cluster_landfall_counts,
    'Total Storms': cluster_total_counts,
    'Landfall Ratio': cluster_landfall_ratio
}).reset_index()

landfall_stats_df
```

```python
# Create a list to store max wind speed data for each cluster
cluster_wind_data = []
unique_clusters = np.unique(storms['cluster'].values)

for cluster in unique_clusters:
    cluster_data = storms.where(storms['cluster'] == cluster, drop=True)
    
    # Process each storm's max wind speed
    cluster_wind_list = []
    for storm_wind in cluster_data['usa_wind']:
        # Remove NaNs and find the max wind speed for this storm
        valid_wind = storm_wind[~np.isnan(storm_wind)]
        if len(valid_wind) > 0:
            cluster_wind_list.append(valid_wind.max().values)
    
    # Add to data only if the cluster_wind_list is not empty
    if cluster_wind_list:
        cluster_wind_data.append(cluster_wind_list)

# Make boxplots
plt.figure(figsize=(10, 6))
plt.boxplot(cluster_wind_data, labels=[f'Cluster {c}' for c in unique_clusters if len(cluster_wind_data[unique_clusters.tolist().index(c)]) > 0])
plt.title('Maximum Wind Speed Distribution by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Maximum Wind Speed (knots)')
plt.show()
```

```python
ws_means = storms.ws.groupby("cluster").mean(dim=("storm", "date_time"))
print(ws_means)
```

```python
# Create a list to store WS data for each cluster
cluster_ws_data = []
unique_clusters = np.unique(storms['cluster'].values)

for cluster in unique_clusters:
    cluster_data = storms.where(storms['cluster'] == cluster, drop=True)
    
    # Process NaNs separately for each storm in the cluster
    cluster_ws_list = []
    for storm_ws in cluster_data['ws']:
        # Remove NaN values in this storm's wind shear data
        valid_ws = storm_ws[~np.isnan(storm_ws)]
        
        # Add to the list only if valid data exists
        if len(valid_ws) > 0:
            cluster_ws_list.extend(valid_ws.values)
    
    # Append the cluster list only if it's non-empty
    if cluster_ws_list:
        cluster_ws_data.append(cluster_ws_list)

# Create boxplots for wind shear distribution by cluster
plt.figure(figsize=(10, 6))
plt.boxplot(cluster_ws_data, labels=[f'Cluster {c}' for c in unique_clusters if len(cluster_ws_data[unique_clusters.tolist().index(c)]) > 0])
plt.title('Wind Shear Distribution by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Wind Shear')
plt.show()
```

```python
# Clusters with low median wind shear (e.g., Cluster 2) 
# are more likely to consist of hurricanes that strengthened significantly.

# Clusters with high median wind shear (e.g., Clusters 0, 1)
# likely include hurricanes that struggled to intensify or weakened.
```
